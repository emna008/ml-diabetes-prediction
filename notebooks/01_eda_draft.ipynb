# Draft 01 — Analyse exploratoire

**Projet :** Prédiction du diabète (classification binaire)

**Dataset :** Pima Indians Diabetes Database (UCI / Kaggle)

```python
import sys
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

ROOT = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.insert(0, str(ROOT / 'src'))

from preprocessing import FEATURE_COLUMNS, TARGET_COLUMN, load_data

sns.set_theme(style='whitegrid')
df = load_data(ROOT / 'data' / 'diabetes.csv')
df.head()
```

## Question métier

Peut-on prédire la présence de diabète (`Outcome = 1`) à partir de mesures cliniques comme la glycémie, l'IMC et l'âge ?

```python
print(f'Nombre de lignes: {len(df)}')
print(f'Nombre de colonnes: {df.shape[1]}')
df.info()
df.describe().T
```

```python
missing_zeros = df.isna().sum().sort_values(ascending=False)
print('Valeurs manquantes après remplacement des zéros impossibles:')
print(missing_zeros[missing_zeros > 0])

class_counts = df[TARGET_COLUMN].value_counts()
print('\nDistribution de la cible:')
print(class_counts)
print(class_counts / len(df))
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
sns.countplot(data=df, x=TARGET_COLUMN, ax=axes[0])
axes[0].set_title('Distribution des classes')
sns.heatmap(df[FEATURE_COLUMNS + [TARGET_COLUMN]].corr(), annot=True, fmt='.2f', ax=axes[1])
axes[1].set_title('Matrice de corrélation')
plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 4, figsize=(16, 8))
axes = axes.ravel()
for i, col in enumerate(FEATURE_COLUMNS):
    sns.boxplot(data=df, x=TARGET_COLUMN, y=col, ax=axes[i])
    axes[i].set_title(col)
plt.tight_layout()
plt.show()
```

## Hypothèses initiales

- La **glycémie (Glucose)** et l'**IMC (BMI)** sont fortement associées au diabète.
- Le dataset est **déséquilibré** (~65% non-diabétiques).
- Des zéros physiologiquement impossibles ont été remplacés par des valeurs manquantes.
